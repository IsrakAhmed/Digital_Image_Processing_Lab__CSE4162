1.	Take grayscale image of size 512x512 and perform the following operations:

    (b)	Decrease it intensity level resolution by one bit up to reach its binary format observe its change when displaying in the same window size.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


image = cv.imread('img2.jpg', cv.IMREAD_GRAYSCALE)
image = cv.resize(image, (512, 512))


plt.figure(figsize=(16,10))


def decreasedResolution(image, numOfBits):
    # Calculate the maximum pixel value for the given bit depth
    maxPixelValue = 2 ** numOfBits - 1
    
    # Reduce the intensity resolution
    # maxPixelValue / 255 calculates the scaling factor
    decreasedImage = (image * (maxPixelValue / 255)).astype(np.uint8)
    return decreasedImage


for i in range(1, 9):
    decreasedImage = decreasedResolution(image, 9-i)
    plt.subplot(2, 4, i)
    plt.imshow(decreasedImage, cmap='gray')
    plt.axis('off')
    plt.title(f'Bits: {9-i}')
```
