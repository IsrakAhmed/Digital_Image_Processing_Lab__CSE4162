### 2.	Take grayscale image of size 512x512 and perform the following operations:

        (b)	Differentiate the results of power law transform and inverse logarithmic transform.

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
originalImage = cv2.imread("img.jpg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

### Power Law Transform

```python
# Power-law Transform for gamma: T(r) = c * r^gamma

def powerLawTransform(image, gamma):   
    
    c = 1

    return c * image ** gamma
```

### Logarithmic Transform

```python
# Logarithmic Transform: T(r) = c * log(1 + r)

def logTransform(image):

    c = np.max(image) / np.log(1 + np.max(image))
    
    return c * np.log(1 + image)  # T(x)=c * log(1+x)
```

### Inverse Logarithmic Transform

```python
# Inverse Logarithmic Transform: T(r) = c * (exp(r) - 1)

def inverseLogTransform(image):

    c = 255 / (np.exp(1) - 1) 

    #return c * np.expm1(image / 255)    # T^−1 (x) = c * exp(x)−1

    return np.expm1(image / c)    # T^−1 (x) = exp(x / c)−1
```

### Plot All Images

```python
plt.figure(figsize=(12, 24))


plt.subplot(4, 2, 1)
plt.imshow(originalImage, cmap='gray')
plt.axis('off')
plt.title(f"Original Image")


powerLawTransformedImage = powerLawTransform(originalImage, 0.4)
plt.subplot(4, 2, 2)
plt.imshow(powerLawTransformedImage, cmap='gray')
plt.axis('off')
plt.title(f"Power-law Transformed Image (gamma=0.4)")


logTransformedImage = logTransform(originalImage)
plt.subplot(4, 2, 3)
plt.imshow(logTransformedImage, cmap='gray')
plt.axis('off')
plt.title(f"Logarithmic Transformed Image")


inverseLogTransformedImage = inverseLogTransform(originalImage)
plt.subplot(4, 2, 4)
plt.imshow(inverseLogTransformedImage, cmap='gray')
plt.axis('off')
plt.title(f"Inverse Logarithmic Transformed Image")
```
