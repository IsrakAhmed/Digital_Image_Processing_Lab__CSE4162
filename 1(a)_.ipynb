1.	Take grayscale image of size 512x512 and perform the following operations:

    (a)	Decrease its spatial resolution by half every time and observe its change when displaying in the same window size.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('img2.jpg', cv.IMREAD_GRAYSCALE)
image = cv.resize(image, (512, 512))

plt.figure(figsize=(15,10))

def display_image(img, title, ax):
    ax.imshow(img, cmap='gray')
    ax.title(title)
    ax.axis('off')


plt.subplot(2, 2, 1)
display_image(image, "Original (512 x 512)", plt)

for i in range (2, 5):

    #image = cv.pyrDown(image) #same as cv.resize()

    # Calculate new size: half the width and half the height
    new_size = (image.shape[1] // 2, image.shape[0] // 2)

    # Resize the image to the new size
    image = cv.resize(image, new_size, interpolation=cv.INTER_AREA)

    plt.subplot(2, 2, i)

    display_image(image, f"Resolution ({image.shape[1]} x {image.shape[0]})", plt)

plt.tight_layout()
plt.show()
```
