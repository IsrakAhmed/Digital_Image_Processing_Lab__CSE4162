### 2.	Take grayscale image of size 512x512 and perform the following operations: 
        
        (a) Perform the brightness enhancement of a specific range of gray levels and observe its result.


```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
originalImage = cv2.imread("img.jpg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

### Define the range of gray levels to enhance

```python
lower_bound = 50  # Lower bound of the range
upper_bound = 150  # Upper bound of the range
brightness_increase = 50  # Amount to increase brightness
```

### Create a copy of the image to modify

```python
enhanced_image = originalImage.copy()
```

### Apply brightness enhancement to pixels within the range

```python
mask = (enhanced_image >= lower_bound) & (enhanced_image <= upper_bound)
enhanced_image[mask] = np.clip(enhanced_image[mask] + brightness_increase, 0, 255)
```

### Visualize the results

```python
plt.figure(figsize=(10, 5))

# Display original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(originalImage, cmap='gray')
plt.axis('off')

# Display enhanced image
plt.subplot(1, 2, 2)
plt.title("Enhanced Image")
plt.imshow(enhanced_image, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
