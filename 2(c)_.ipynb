### 2.	Take grayscale image of size 512x512 and perform the following operations:

        (c)	Find the difference image between original and the image obtained by last three bits (MSB).


```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
originalImage = cv2.imread("img.jpg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
msbImage = originalImage & 0b11100000

differenceImage = abs(originalImage - msbImage)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(originalImage, cmap='gray')
plt.title(f"Original Image")

plt.subplot(1, 3, 2)
plt.imshow(msbImage, cmap='gray')
plt.title(f"MSB Image")

plt.subplot(1, 3, 3)
plt.imshow(differenceImage, cmap='gray')
plt.title(f"Difference Image")
```
